```python
import sklearn
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=500, n_features=2, centers=[(-1,-0.5), (0.5, -0.5)], cluster_std=0.5)
print(X.shape) 
fig, ax = plt.subplots()
ax.plot(X[y==0, 0], X[y==0, 1], "x", color='blue')
ax.plot(X[y==1, 0], X[y==1, 1], "*", color='green')
fig.show()
```
